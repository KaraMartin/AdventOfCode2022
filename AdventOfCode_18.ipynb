### Day 18 - Boiling Boulders
##### Part 1 - Total Surface Area

What is the surface area of your scanned lava droplet?

```python
def part1(input_file):
    with open(input_file, "r") as f:
        data = [tuple(map(int, a.split(','))) for a in f.read().strip().split("\n")] 

    boxes = set(data)
    xyz = list(zip(*data))
    res = 0
    dirs = ((1, 0, 0), (0, 1, 0), (0, 0, 1))

    for x,y,z in data:
        free_sides = 6
        for d in dirs:
            a, b, c, aa, bb, cc = x + d[0], y + d[1], z + d[2], x - d[0], y - d[1], z - d[2]
            free_sides -= ((a, b, c) in boxes) + ((aa, bb, cc) in boxes)
        res += free_sides

    return res
```

```python
print(f"Part 1: Surface area of scanned lava droplet is {part1('aoc_18_input.txt')}")             
```

Part 2 - Exterior Surface Area

Find only the exterior surface area of the scanned lava droplet.

```python
'''s = """2,2,2
1,2,2
3,2,2
2,1,2
2,3,2
2,2,1
2,2,3
2,2,4
2,2,6
1,2,5
3,2,5
2,1,5
2,3,5"""
'''
with open("aoc_18_input.txt", "r") as f:
    s = f.read().strip()

data = [tuple(map(int, a.split(','))) for a in s.split("\n")]
boxes = set(data)
xyz = list(zip(*data))
minx, maxx, miny, maxy, minz, maxz = min(xyz[0]), max(xyz[0]), min(xyz[1]), max(xyz[1]), min(xyz[2]), max(xyz[2])    
print(f"{minx, maxx, miny, maxy, minz, maxz}")
dirs = ((1, 0, 0), (0, 1, 0), (0, 0, 1))

# floodfill on outside
stack = [(1,1,1)]
outside = set()
outside.add((1,1,1))

while stack:
    curr = stack.pop()
    for d in dirs:
        xx,yy,zz = curr[0] + d[0], curr[1] + d[1], curr[2] + d[2]
        if all([xx >= minx - 1, yy >= miny - 1, zz >= miny-1, xx <= maxx+1, yy <= maxy+1, zz <= maxz+1, (xx,yy,zz) not in boxes, (xx,yy,zz) not in outside]):
            stack.append((xx,yy,zz))
            outside.add((xx,yy,zz))
        xx,yy,zz = curr[0] - d[0], curr[1] - d[1], curr[2] - d[2]
        if all([xx >= minx - 1, yy >= miny - 1, zz >= miny-1, xx <= maxx+1, yy <= maxy+1, zz <= maxz+1, (xx,yy,zz) not in boxes, (xx,yy,zz) not in outside]):
            stack.append((xx,yy,zz))
            outside.add((xx,yy,zz))

res = 0
for x,y,z in sorted(data, key = lambda x : (x[2], x[0], x[1])):
    cnt = 0
    for d in dirs:
        a, b, c, aa, bb, cc = x + d[0], y + d[1], z + d[2], x - d[0], y - d[1], z - d[2]
        cnt += ((a, b, c) in outside) + ((aa, bb, cc) in outside)
    res += cnt

print(f"{res}")
```
